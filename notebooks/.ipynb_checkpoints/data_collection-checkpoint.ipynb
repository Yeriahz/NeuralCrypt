```python
# Import Libraries
import requests
import pandas as pd
import time

print("Libraries imported successfully!")

#Replace 'your_api_key_here' with your actual API key from Cryptocompare
API_KEY = '9bf2ff68d6680c3f789283da46195442cef9ee8f601182cfc731f248ab6616e9'
BASE_URL = 'https://min-api.cryptocompare.com/data/v2/'

print("API key and base URL set up.")

#Function to fetch data from any endpoint
def fetch_data(endpoint, params):
    url = f"{BASE_URL}{endpoint}"
    params["api_key"] = API_KEY
    response = requests.get(url, params=params)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error fetching data from {endpoint}: {response.status_code}")
        return None

#Historical Daily Data
def get_historical_daily(crypto="BTC", currency="USD", limit=30):
    params = {"fsym": crypto, "tsym": currency, "limit": limit}
    data = fetch_data("histoday", params)
    if data:
        return pd.DataFrame(data["Data"]["Data"])
    return None

#Historical Hourly Data
def get_historical_hourly(crypto="BTC", currency="USD", limit=30):
    params = {"fsym": crypto, "tsym": currency, "limit": limit}
    data = fetch_data("histohour", params)
    if data:
        return pd.DataFrame(data["Data"]["Data"])
    return None

#Historical Minute Data
def get_historical_minute(crypto="BTC", currency="USD", limit=30):
    params = {"fsym": crypto, "tsym": currency, "limit": limit}
    data = fetch_data("histominute", params)
    if data:
        return pd.DataFrame(data["Data"]["Data"])
    return None

#Real-Time Price Data
def get_real_time_price(crypto="BTC", currency="USD"):
    url = f"https://min-api.cryptocompare.com/data/price"
    params = {"fsym": crypto, "tsyms": currency, "api_key": API_KEY}
    response = requests.get(url, params=params)
    return response.json() if response.status_code == 200 else None

#Top Cryptocurrencies by Market Cap
def get_top_market_cap(limit=10):
    params = {"limit": limit, "tsym": "USD"}
    data = fetch_data("top/mktcapfull", params)
    if data:
        return pd.DataFrame([item["CoinInfo"] for item in data["Data"]])
    return None

# Main Function
def main():
    print("Fetching historical daily data...")
    daily_data = get_historical_daily("BTC", "USD", 30)
    if daily_data is not None:
        daily_data.to_csv("btc_historical_daily.csv", index=False)
        print("Saved daily data to btc_historical_daily.csv")

    print("Fetching historical hourly data...")
    hourly_data = get_historical_hourly("BTC", "USD", 30)
    if hourly_data is not None:
        hourly_data.to_csv("btc_historical_hourly.csv", index=False)
        print("Saved hourly data to btc_historical_hourly.csv")

    print("Fetching historical minute data...")
    minute_data = get_historical_minute("BTC", "USD", 30)
    if minute_data is not None:
        minute_data.to_csv("btc_historical_minute.csv", index=False)
        print("Saved minute data to btc_historical_minute.csv")

    print("Fetching real-time price data...")
    real_time_price = get_real_time_price("BTC", "USD")
    if real_time_price:
        print("Real-Time BTC Price:", real_time_price)

    print("Fetching top cryptocurrencies by market cap...")
    top_cryptos = get_top_market_cap(10)
    if top_cryptos is not None:
        top_cryptos.to_csv("top_cryptos.csv", index=False)
        print("Saved top market cap data to top_cryptos.csv")

#Run the script
if __name__ == "__main__":
    main()
```
